# Цель на данной практической работе необходимо использовать датасет с прошлой практической работы и визуализировать данные в виде графиков. 

Для начала импортируем Pandas для прочтения датасета

```python
import pandas as panda
overwatch = panda.read_csv('Overwatch_edopic.csv')
overwatch
```

Далле импортируем Matplotlib (он нам понадобится для черчения графиков и т.п)

```python
import matplotlib.pyplot as mat
```

Теперь нам необходимо сделать Линейные графики (в моём случае сделаю 3)

```python
mat.plot(overwatch['Heroes'].head(5), overwatch['Life'].head(5), label='Хп', color = "Red")
mat.title('Статистика героев по хп')
mat.xlabel('Имя героя')
mat.ylabel('Хп героя')
mat.legend()
mat.show()
```

Здесь мы построили линейный график соотношения героев и их хп. На данном графике мы не указывали маркер, выбрали красный цвет самого графика. Выбрали заголовок, указали наименования по X и Y. Использовали head. legend показывает легенду, а show выводит график.

```python
mat.plot(overwatch['Heroes'].tail(20), overwatch['Health per Sec'].tail(20), label='Хил', alpha = 0.5, marker = 'x')
mat.title('Статистика героев по хилу')
mat.xlabel('Имя героя')
mat.ylabel('Хил героя')
mat.legend()
mat.show()
```

Здесь мы построили линейный график соотношения героев и их хила. На данном графике мы указали маркер, оставили дефолтный цвет графика, но понизили его альфу. Выбрали заголовок, указали наименования по X и Y. Использовали tail. legend показывает легенду, а show выводит график.

```python
mat.plot(overwatch['Heroes'].head(5), overwatch['Reload'].head(5), label='Перезарядка', color = "Cyan", alpha = 0.5, marker = "o")
mat.title('Статистика героев по перезарядке')
mat.xlabel('Имя героя')
mat.ylabel('Перезарядка героя')
mat.legend()
mat.show()
```

Здесь мы построили линейный график соотношения героев и их хила. На данном графике мы указали маркер, поменяли цвет графика и измениили его альфу. Выбрали заголовок, указали наименования по X и Y. Использовали head. legend показывает легенду, а show выводит график.

Далее мы сделаем 3 столбчатые диаграммы

```python
mat.bar(overwatch['Heroes'].head(5), overwatch['Health per Sec'].head(5), label='Хил')
mat.title('Статистика героев по хилу')
mat.xlabel('Имя героя')
mat.ylabel('Хил героя')
mat.legend()
mat.show()
```

Здесь столбчатый график соотношения героев к его хилу. Мы не меняли его цвет. Маркеры здесь не нужны. Выбрали заголовок, указали наименования по X и Y. Использовали head. legend показывает легенду, а show выводит график.

```python
mat.bar(overwatch['Heroes'].tail(5), overwatch['Reload'].tail(5), label='Перезарядка', color = "Green")
mat.title('Статистика героев по перезарядке')
mat.xlabel('Имя героя')
mat.ylabel('Перезарядка героя')
mat.legend()
mat.show()
```

Здесь столбчатый график соотношения героев к его перезарядке. Поменяли его цвет на зеленый. Маркеры здесь не нужны. Выбрали заголовок, указали наименования по X и Y. Использовали tail. legend показывает легенду, а show выводит график.

```python
mat.bar(overwatch['Heroes'].head(5), overwatch['Life'].head(5), label='Хп', color = "Purple", alpha = 0.5)
mat.title('Статистика героев по хп')
mat.xlabel('Имя героя')
mat.ylabel('Хп героя')
mat.legend()
mat.show()
```

Здесь столбчатый график соотношения героев к его перезарядке. Поменяли его цвет на фиолетовый и изменили альфу. Маркеры здесь не нужны. Выбрали заголовок, указали наименования по X и Y. Использовали tail. legend показывает легенду, а show выводит график.

Далее сделаем Круговые диаграммы

```python
data = overwatch['Health per Sec'].value_counts().head(5)
mat.pie(data, labels=data.index, autopct='%0.0f%%', colors = ['Gray','Green'])
mat.legend()
mat.show()
```

Здесь круговая диаграмма соотношения героев к его хилу. Мы выбрали ему серый и зеленый цвета. Маркеры здесь не нужны. Здесь мы указали новую переменную data с хранением нужных нам данных. autopct использовали для указания формата. Использовали head. legend показывает легенду, а show выводит график.

```python
data = overwatch['Reload'].value_counts().tail(5)
mat.pie(data, labels=data.index, colors = ['Red', 'Yellow'])
mat.legend()
mat.show()
```

Здесь круговая диаграмма соотношения героев к его перезарядке. Мы выбрали ему желтый и красные цвета. Маркеры здесь не нужны. Здесь мы указали новую переменную data с хранением нужных нам данных. autopct не использовали. Использовали tail. legend показывает легенду, а show выводит график.

Теперь сделаем диаграммы рассеяния

```python
mat.scatter(overwatch['Health per Sec'], overwatch['Reload'], alpha = 0.5)
mat.show()
```

Здесь диаграмма рассеяния соотношения героев к его хилу. Мы не выбирали ему цвет, но измениили альфу. Маркеры здесь не нужны. show выводит график.

```python
mat.scatter(overwatch['Heroes'].head(5), overwatch['Life'].head(5), color = "Black")
mat.show()
```

Здесь диаграмма рассеяния соотношения героев к его хилу. Мы выбрали ему черный цвет. Маркеры здесь не нужны. Использовали head, show выводит график.

Теперь изобразим гистограмму

```python
mat.hist(overwatch['Health per Sec'], bins = 5, label = 'Хил в секунду', color = 'Pink')
mat.legend()
mat.show()
```

Здесь гистограмма хила. Мы выбрали ей розовый цвет. Маркеры здесь не нужны. Создали мы её с помощью фунции hist, и задали кол-во столбцов через bins. legend показывает легенду, show выводит график.

```python
mat.hist(overwatch['Nanoboost DPS'], bins = 10, label = 'Нанобусты', alpha = 0.5)
mat.legend()
mat.show()
```

Здесь гистограмма нанобуста. Мы оставили дефолтный цвет и указали альфу. Маркеры здесь не нужны. Создали мы её с помощью фунции hist, и задали кол-во столбцов через bins. legend показывает легенду, show выводит график.

```python
mat.hist(overwatch['Single shot'], bins = 10, label = 'Синглшоты', color = 'red', alpha = 0.5)
mat.legend()
mat.show()
```

Здесь гистограмма синглшотов. Мы выбрали ей красный цвет и указали альфу. Маркеры здесь не нужны. Создали мы её с помощью фунции hist, и задали кол-во столбцов через bins. legend показывает легенду, show выводит график.
